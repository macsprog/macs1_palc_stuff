## Rotation dans le plan

On considre:
* un point du plan $(x,y)$
* rotation d'un point autour d'un point $(x_0,y_0)$ par un angle $\theta$

\begin{align*}
x& = x_0 + \cos(\theta)(x-x_0) - \sin(\theta)(y-y_0)
\\
y&= y_0 + \sin(\theta)(x-x_0) + \cos(\theta)(y-y_0)
\end{align*}

## choix d'implmentation

* $(x,y)$ reprsent par deux flottants `x` et `y`
* crire une fonction pour raliser la rotation
